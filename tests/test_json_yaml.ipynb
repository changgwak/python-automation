```python
import yaml
import os

def parse_data(input=None, **kwargs) -> dict:
    # Default parameters
    data = {'key': 1, 'another_key': 'another_default_value'}

    if input is not None:
        # Check if input is a file path
        if os.path.isfile(input):
            with open(input, 'r') as file:
                file_data = yaml.safe_load(file)
                print(type(file_data), file_data)
                data.update(file_data)
        else:
            try:
                yaml_data = yaml.safe_load(input)
                print(type(yaml_data), yaml_data)
                data.update(yaml_data)
            except yaml.YAMLError:
                print("Invalid YAML data")
                return None
    else:
        data.update(kwargs)

    return data


# Call the function with a file path
data = parse_data(r'sample.yaml')
print(data)

# Call the function with individual arguments
data = parse_data(key=3, another_key="another_new_value")
print(data)

# Call the function with individual arguments
data = parse_data(key=4)
print(data)

# Call the function with no arguments to get default parameters
data = parse_data()
print(data)


## sample.yaml
# key: yaml_file_value
# another_key: yaml_file_another_value
# extra_key: extra_value
```

```python


import json
import os

def parse_data(input=None, **kwargs) -> dict:
    # Default parameters
    data = {'key': 1, 'another_key': 'another_default_value'}
    # print(type(data))
    if input is not None:
        # Check if input is a file path
        if os.path.isfile(input):
            with open(input, 'r') as file:
                file_data = json.load(file)
                data.update(file_data)
        else:
            try:
                json_data = json.loads(input)
                data.update(json_data)
            except json.JSONDecodeError:
                print("Invalid JSON data")
                return None
    else:
        data.update(kwargs)

    return data

data = parse_data(r'sample.json')
print(data)

# Call the function with a JSON string or file path
data = parse_data('{"key": 2}')
print(data)

# Call the function with individual arguments
data = parse_data(key=3, another_key="another_new_value")
print(data)

# Call the function with individual arguments
data = parse_data(key=4)
print(data)

# Call the function with no arguments to get default parameters
data = parse_data()
print(data)


## sample.json
# {
#     "key": "json_file_value",
#     "another_key": "json_file_another_value",
#     "extra_key": "extra_value"
# }
```
